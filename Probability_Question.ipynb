3 data scientists work on a cluster which has a daily limit of 2 compute ’credits’. When the total daily amount of compute credits of the three data scientists exceeds that daily limit, the company is charged a one thousand dollar penalty. Assume each data scientist’s daily cluster usage in compute credits is distributed between 0 and 1 and that the chance the usage belongs in some subinterval of [0, 1] is equal to the length of the subinterval. Assume also that the daily usages of the data scientists across the year are independent. You are asked to estimate the total cost of this penalty over one business year (252 days) with a single number. What is your succinct response?

```
import numpy as np

one = np.random.uniform(0,1)
two = np.random.uniform(0,1)
three = np.random.uniform(0,1)
```

```
print(one, two, three)
```

```
def check_usage():
  """
  Uniformly generate the usage using numpy package 
  """
  total_usage = []
  for i in range(252):
    one = np.random.uniform(0,1)
    two = np.random.uniform(0,1)
    three = np.random.uniform(0,1)
    total_usage.append(one + two + three)
  return total_usage
```

```
total_usage = check_usage()

threshold_breach = [i for i in total_usage if i >2]

print('The number of days usage of 3 scientists crossed the threshold of 2 compute credits: ', int(sum(threshold_breach)))
print('Total cost of the breach : ', '$', int(sum(threshold_breach))*1000, )

import matplotlib.pyplot as plt
plt.plot(total_usage)
plt.axhline(y=2, color='r')
plt.show()
```
